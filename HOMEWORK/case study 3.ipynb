```python
class Portfolio:
    def __init__(self, name, size, roe, price):
        self.name = name
        self.size = size
        self.roe = roe
        self.price = price
from tkinter import *
from tkinter import messagebox
from tkinter import ttk

window = Tk()
window.title('MY PORTFOLIO ^_^')
window.configure(bg='light blue')
list_portfolio = [] #Luutru danh sach cac ma co phieu

window.geometry('750x450')
lbl = Label(window, text="Earning money for fun!", bg='light blue', relief=FLAT, width=40, height=1,font=("Arial Bold", 20))
lbl.grid(column=0, row=0, columnspan=3)
s = ttk.Style()
s.theme_use('clam')

name_lbl = Label(window, text="Name: ", bg='light blue', font=("Arial Bold", 12), padx=10, pady=10)
size_lbl = Label(window, text="Size (bil): ", bg='light blue', font=("Arial Bold", 12), padx=10, pady=10)
roe_lbl = Label(window, text="ROE: ", bg='light blue', font=("Arial Bold", 12), padx=10, pady=10)
price_lbl = Label(window, text="Price (VND): ", font=("Arial Bold", 12), bg='light blue', padx=10, pady=10)
lst_portfolio_lbl = Label(window, text="PORTFOLIO", font=("Arial Bold", 15), padx=15, pady=15)

name = Entry(window)
size = Entry(window)
roe = Entry(window)
price = Entry(window)
check = Entry(window)

def add_stock():
    # cach luu thong tin tu entry
    list_portfolio.append(Portfolio(name.get(), str(size.get()), str(roe.get()), str(price.get())))
    print([item.name for item in list_portfolio])
    messagebox.showinfo("Message", f"{name.get()} added successfully.")
    lst_portfolio.insert(END, list_portfolio[len(list_portfolio)-1].name + '   ' + list_portfolio[len(list_portfolio)-1].size + '   ' + list_portfolio[len(list_portfolio)-1].roe + '   ' + list_portfolio[len(list_portfolio)-1].price)
    

def search_stock():
    search_name = check.get()
    exist = False # gia dinh la item khong ton tai trong list
    for item in list_portfolio:
        if item.name == search_name: # da co ket qua tim kiem
            exist = True # item dang tim kiem co ton tai trong list
            messagebox.showinfo("Portfolio", f"{item.name} has price VND {item.price}, with ROE: {item.roe} and size of company bil: {item.size} .")
            break
    if exist == False:
            messagebox.showinfo("Portfolio", f"{search_name} does not exist in the list.") 

def delete_stock():
    delete_name = check.get()
    exist = False
    for i in range (len(list_portfolio)):
        if list_portfolio[i].name == delete_name:
            exist = True
            del list_portfolio[i]
            messagebox.showinfo("Portfolio", f"{delete_name} is deleted successfully.") 
            break
    if exist == False:
        messagebox.showinfo("Portfolio", f"{delete_name} does not exist in the list.") 
            
def reset_stock():
    name.delete(0, len(name.get()))
    size.delete(0, len(size.get()))
    roe.delete(0, len(roe.get()))
    price.delete(0, len(price.get()))

def edit_stock():
    edit = check.get()
    exist = False
    for item in range(len(list_portfolio)):
        if list_portfolio[item].name == check:
            list_portfolio[item].name == name.get()
            list_portfolio[item].size = size.get()
            list_portfolio[item].roe = roe.get()
            list_portfolio[item].price = price.get()
            del list_portfolio[item]
            for item in {len(list_portfolio) - 1}:
                lst_portfolio.insert(parent='',index='end',values=(item.name, item.size, item.roe, item.price))
            messagebox.showinfo("Portfolio", f"Update information:{list_portfolio[item].name} has price VND {list_portfolio[item].price}, with ROE: {list_portfolio[item].roe} and size of company bil: {list_portfolio[item].size}.")
            exit = True
            break
    if exist == False:
        messagebox.showinfo("Portfolio", f"{check} không tìm được trong danh sách")
  
# map nut button voi ham
add_btn = Button(window, text="Add", bg="pink", fg="white", command=add_stock)
delete_btn = Button(window, text="Delete", bg="white", fg="red", command=delete_stock)
search_btn = Button(window, text="Search",bg="red", fg="yellow", command=search_stock)
reset_btn = Button(window, text="Reset", bg="yellow", fg="green", command=reset_stock)
edit_btn = Button(window, text="Edit", bg="green", fg="pink", command=edit_stock)

# tao label cho cac thuoc tinh
name_lbl.grid(row=1, column=0)
size_lbl.grid(row=2, column=0)
roe_lbl.grid(row=3, column=0)
price_lbl.grid(row=4, column=0)
lst_portfolio_lbl.grid(row=11, column=1)
name.grid(row=1, column=1)
size.grid(row=2, column=1)
roe.grid(row=3, column=1)
price.grid(row=4, column=1)

# tao label cho cac button
add_btn.grid(row=1, column=2)
delete_btn.grid(row=2, column=2)
reset_btn.grid(row=3, column=2)
edit_btn.grid(row=4, column=2)
search_btn.grid(row=5, column=2)
check.grid(row=5, column=1)

# tao giao dien the hien list
lst_portfolio = Listbox(window, selectmode=MULTIPLE, relief=SUNKEN, width=100, height=50)
lst_portfolio_lbl.place(x= 300, y=250)
lst_portfolio.place(x=70, y=300)

window.mainloop()
```
